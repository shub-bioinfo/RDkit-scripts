```python
#Question 4
ben = Chem.MolFromSmiles('C1=CC=CC=C1')
pyr = Chem.MolFromSmiles('C1=CC=NC=C1')
ben
```

```python
pyr
```

```python

bz={}
py={}
#Getting their Finger prints
benfp= AllChem.GetMorganFingerprintAsBitVect(ben, 2, nBits=1024, bitInfo=bz)
pyrfp= AllChem.GetMorganFingerprintAsBitVect(pyr, 2, nBits=1024, bitInfo=py)
```

```python
Tan = DataStructs.TanimotoSimilarity(pyrfp,benfp)
print(Tan)
```

```python
print('Ben', list(benfp.GetOnBits()))
print('pyr', list(pyrfp.GetOnBits()))
```
